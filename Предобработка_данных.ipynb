##### Заполним пропуски в длительности фильма

```python
df[(df['minutes'].isna() == False)&(df['hours'].isna() == True)]
```

```python
df['hours'] = df.apply(lambda row: 0 if row['hours']!=row['hours'] and row['minutes'] > 0 else row['hours'], axis = 1)
```

```python
df['minutes'] = df.apply(lambda row: 0 if row['minutes']!=row['minutes'] and row['hours'] > 0 else row['minutes'], axis = 1)
```

```python
df['duration_min'] = df['hours']*60+df['minutes']
```

```python
df.sample(5)
```

```python
df.info()
```

```python
df['duration_min'].describe()
```

```python
df.query('duration_min > 17000')
```

```python
means = df.groupby(['type', 'episodes'])['duration_min'].mean()
```

```python
means = means.reset_index()
```

```python
means
```

```python
df['episodes'] = df['episodes'].replace(0, 1).fillna(1)
```

```python
avg_duration = df[df['duration_min'] >0]['duration_min'].median()
avg_duration
```

```python
def dur(duration, type, episodes):
  if duration == 0 or duration != duration:
   try:
    # подставляем значение из датафрейма means
    return(means[(means['type']==type)&(means['episodes']==episodes)]).iloc[0]['duration_min']
   except:
    # если такого значения нет (например, все нули), подставляем среднее по всем фильмам
    return avg_duration
  else:
    # если длительность >0 оставляем это значение
    return duration
```

```python
df['duration_min'] = df.apply(lambda x: dur(x['duration_min'], x['type'], x['episodes']),axis=1)
```

Стемминг

```python
from nltk.stem import SnowballStemmer 
russian_stemmer = SnowballStemmer('russian')  
```

```python
russian_stemmer.stem('счастье')
```

```python
df['name'] = df['name'].str.lower()
```

```python
list_stem = []
for movie in df['name']:
    for word in movie.split():
        stemmed_word = russian_stemmer.stem(word)
        if stemmed_word == 'счаст':
            list_stem.append(movie)
            print(movie) 
print(len(list_stem)) 
```

Лемматизация

```python
from pymystem3 import Mystem
m = Mystem() 
```

```python
m.lemmatize('любовь')
```

```python
list_lemm = []
for movie in df['name']:
    for lemma in m.lemmatize(movie):
        if lemma == 'любовь':
            list_lemm.append(movie)
print(len(list_lemm)) 
```
