## Функции

Добавить столбец is_perfect, который равен True, если value - совершенное число и False - если нет.

Совершенное число — натуральное число, равное сумме всех своих собственных делителей (то есть всех положительных делителей, отличных от самого числа). Например, число 6 равно сумме своих собственных делителей 1 + 2 + 3.

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({"index":[1, 2, 3, 4, 5], "value":[6, 14, 28, 15, 48]})
df
```

```python
def is_perfect(n):
    s = 0
    for i in range(1,n):
         if n % i == 0:
            s += i
    return n == s
```

```python
is_perfect(4)
```

```python
df['is_perfect'] = df['value'].apply(is_perfect)
```
